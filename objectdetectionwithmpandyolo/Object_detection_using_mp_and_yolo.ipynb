```python
import os
import cv2
import mediapipe as mp
import time

# Set CUDA_VISIBLE_DEVICES to an empty string to force CPU execution
os.environ['CUDA_VISIBLE_DEVICES'] = ''

mp_objectron = mp.solutions.objectron
mp_drawing = mp.solutions.drawing_utils

# Specify the path to your video file
video_path = r"/content/drive/MyDrive/movement /lefttoright.mp4"

cap = cv2.VideoCapture(video_path)

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_file = 'output8.avi'

# Get original video properties
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

out = cv2.VideoWriter(output_file, fourcc, fps, (width, height))

start_time = 0 # Start time in seconds
end_time =  10 # End time in seconds

# Calculate the frame indices corresponding to the start and end times
start_frame = int(start_time * fps)
end_frame = int(end_time * fps)

with mp_objectron.Objectron(static_image_mode=False,
                            max_num_objects=2,
                            min_detection_confidence=0.5,
                            min_tracking_confidence=0.8,
                            model_name='Cup') as objectron:
    frame_index = 0
    while cap.isOpened():
        success, image = cap.read()
        if not success:
            print("Failed to read from video file.")
            break

        if frame_index < start_frame:
            # Skip frames until reaching the start time
            frame_index += 1
            continue
        elif frame_index > end_frame:
            # Stop processing frames after the end time
            break

        start = time.time()
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = objectron.process(image)

        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        if results.detected_objects:
            for detected_object in results.detected_objects:
                mp_drawing.draw_landmarks(image, detected_object.landmarks_2d, mp_objectron.BOX_CONNECTIONS)
                mp_drawing.draw_axis(image, detected_object.rotation, detected_object.translation)
                x = results.detected_objects
                print(x[0].rotation)
                print(x[0].translation)
                print(x[0].scale)

                # Extracting desired information

                print('*')

        end = time.time()
        totalTime = end - start
        fbs = 1 / totalTime


        # Write the frame with annotations to the output video file
        out.write(image)

        frame_index += 1

    # Release the VideoCapture and VideoWriter objects
    cap.release()
    out.release()
```

```python
import cv2
import math
import mediapipe as mp
from ultralytics import YOLO
import numpy as np

mp_objectron = mp.solutions.objectron
mp_drawing = mp.solutions.drawing_utils

# Initialize YOLO model
model_yolo = YOLO("/content/yolov8l.pt")

# Define class names for YOLO
classNames = ["person", "bicycle", "car", "motorbike", "aeroplane", "bus", "train", "truck", "boat",
              "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat",
              "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", "giraffe", "backpack", "umbrella",
              "handbag", "tie", "suitcase", "frisbee", "skis", "snowboard", "sports ball", "kite", "baseball bat",
              "baseball glove", "skateboard", "surfboard", "tennis racket", "bottle", "wine glass", "cup",
              "fork", "knife", "spoon", "bowl", "banana", "apple", "sandwich", "orange", "broccoli",
              "carrot", "hot dog", "pizza", "donut", "cake", "chair", "sofa", "pottedplant", "bed",
              "diningtable", "toilet", "tvmonitor", "laptop", "mouse", "remote", "keyboard", "cell phone",
              "microwave", "oven", "toaster", "sink", "refrigerator", "book", "clock", "vase", "scissors",
              "teddy bear", "hair drier", "toothbrush"]


# Create an instance of the Objectron model specific to cups
objectron = mp_objectron.Objectron(static_image_mode=False,
                                    max_num_objects=2,
                                    min_detection_confidence=0.5,
                                    min_tracking_confidence=0.8,
                                    model_name='Cup')

def convert_to_kitti(sample_data):
    # Extracting data
    rotation = sample_data["rotation"]
    translation = sample_data["translation"]
    scale = sample_data["scale"]

    dimensions = scale
    location = translation
    rotation_y = np.arcsin(-rotation[0, 2])  # Calculating rotation_y

    # Write KITTI format string
    kitti_format = [dimensions[0], dimensions[1], dimensions[2], location[0], location[1], location[2], rotation_y ]

    return kitti_format

def detect_yolo(image):
    results = model_yolo(image, stream=True)
    objects_info = []
    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
            conf = math.ceil((box.conf[0]*100))/100
            cls = int(box.cls[0])
            object_info = {
                "class": classNames[cls],
                "confidence": conf,
                "bounding_box": (x1, y1, x2, y2)
            }
            objects_info.append(object_info)
    return objects_info

def detect_objectron(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image.flags.writeable = False
    results = objectron.process(image)  # Process the image with the Objectron model
    object_info = []
    if results.detected_objects:
        for detected_object in results.detected_objects:
            x = results.detected_objects
            rotation = x[0].rotation
            translation = x[0].translation
            scale = x[0].scale
            object_info.append({
                "rotation": rotation,
                "translation": translation,
                "scale": scale
            })
    return object_info

def process_frame(image):
    yolo_results = detect_yolo(image)

    if not yolo_results:
        return  # No objects detected by YOLO, so return without further processing

    objectron_results = detect_objectron(image)
    l = []
    if objectron_results:
        for i in yolo_results:
            if i['class'] == 'cup':
                 l.append(i['class'])
                 for j in i['bounding_box']:
                     l.append(j)
        for obj_result in objectron_results:
            sample_data = {
                  "rotation": np.array(obj_result['rotation']),
                  "translation": np.array(obj_result['translation']),
                  "scale": np.array( obj_result['scale']),
            }

            kitti_data = convert_to_kitti(sample_data)
            for i in kitti_data:
                l.append(i)
    return l

import os

# Specify the path to your video file
video_path = r"/content/drive/MyDrive/movement /forward.mp4"

# Get total number of frames in the video
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Failed to open video file.")

total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(total_frames)
cap.release()

# Select 100 evenly distributed frames
frames_to_process = np.linspace(0, total_frames - 1, num=10, dtype=np.int64)

# Process selected frames
output = []
cap = cv2.VideoCapture(video_path)
for frame_num in frames_to_process:
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num)
    success, image = cap.read()
    if not success:
        print("Error: Failed to read frame from video.")
        break

    # Process frame
    x = process_frame(image)
    if x is not None:  # Check if any objects were detected
        output.append(x)
    height, width, _ = image.shape
    print("Image Size: {}x{}".format(width, height))


cap.release()
cv2.destroyAllWindows()
```
